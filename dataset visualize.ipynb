```python
train_data=pd.read_csv('./assets/processed dataset/train dataset.csv')
validate_data=pd.read_csv('./assets/processed dataset/validation dataset.csv')
test_data=pd.read_csv('./assets/processed dataset/test dataset.csv')
```

```python
train_data.dtypes
```

```python
train_date_range=pd.date_range(start='1/1/1981',end='2/5/2011')
validate_date_range=pd.date_range(start='2/6/2011',end='7/19/2017')
test_date_range=pd.date_range(start='7/20/2017',end='12/31/2023')
fig,ax=plt.subplots(4,1,sharex=True,figsize=(15,10))
ax[0].plot(train_date_range, train_data["PS"],label="Training Data")
ax[0].plot( validate_date_range,validate_data["PS"],label="Validation Data")
ax[0].plot( test_date_range,test_data["PS"],label="Testing data")
ax[0].set_ylabel("PS")
ax[1].plot(train_date_range, train_data["WS10M"],label="Training Data")
ax[1].plot( validate_date_range,validate_data["WS10M"],label="Validation Data")
ax[1].plot( test_date_range,test_data["WS10M"],label="Testing data")
ax[1].set_ylabel("WS10M")
ax[2].plot(train_date_range, train_data["T2M"],label="Training Data")
ax[2].plot( validate_date_range,validate_data["T2M"],label="Validation Data")
ax[2].plot( test_date_range,test_data["T2M"],label="Testing data")
ax[2].set_ylabel("T2M")
ax[3].plot(train_date_range, train_data["QV2M"],label="Training Data")
ax[3].plot( validate_date_range,validate_data["QV2M"],label="Validation Data")
ax[3].plot( test_date_range,test_data["QV2M"],label="Testing data")
ax[3].set_ylabel("QV2M")
fig.suptitle("Feature plot")
ax[0].legend(loc=(0.7,1.2))
ax[3].set_xlabel("Date")
plt.show()
```
